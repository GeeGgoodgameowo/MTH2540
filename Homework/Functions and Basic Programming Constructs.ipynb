### Pr. 3: Write a function to count the number of even integers in a given vector x. Test your function on: $ (1,3,5,6,8,9) $, seq$(1,100,3)$ and seq$(1, 5001, 7)$.

```python
def numberCount (vector):
  count = 0
  for i in range(len(vector)):
    if vector[i] % 2 == 0:
      count = np.round(count + 1, 3)
  return count

test1 = np.array([1, 3, 5, 6, 8, 9])
test2 = np.arange(1, 100, 3)
test3 = np.arange(1, 5001, 7)


print ("Test 1 equals", numberCount(test1))
print ("Test 2 equals", numberCount(test2))
print ("Test 3 equals", numberCount(test3))
```

### Pr. 4: The sum formula for the finite gemometric series $h(x,n) = \displaystyle \sum_{i=0}^{n} x^i = 1+x+x^2+...+x^n $ is given by $$ h(x,n) = \begin{cases}
\frac{1-x^{n+1}}{1-x} & \mbox {for } x \neq 1 \\
n &\mbox {for } x = 1 \end{cases} $$
### Write a function to compute this formula. Use your function to compute $h(0.7,57)$, $h(4.2,9)$, $h(-1.2,81)$, and $h(1,342)$.
#### **Notes:**
#### a) If ùë• = 1, then the sum is ùëõ.
#### b) You‚Äôll see slight variations of this formula depending on if the lower limit is $ùëñ = 0$ or $ùëñ = 1$ or the upper limit is $ùëõ ‚àí 1$, $ùëõ$, $ùëõ + 1$.
#### c) In calculus II, you will see, or have seen, the formula for the sum of an infinite geometric series $ \displaystyle \sum_{n=0}^{‚àû} a~r^n = a(1 + r + r^2 + r^3 + ...)$ for $ a \neq 0 $. It converges to the sum $ \displaystyle S = \frac{a}{1-r} $ for $|r| < 1$, and diverges for $|r| \geq 1$.


```python
def sumFiniteGeoSeries (x, n):
  if x == 1:
    return n
  else:
    for i in range (n+1):
      sum = np.round((1-(x**(n+1)))/(1-x),3)
      return sum

print ("Test 1 equals", sumFiniteGeoSeries(.7, 57))
print ("Test 2 equals", sumFiniteGeoSeries(4.2, 9))
print ("Test 3 equals", sumFiniteGeoSeries(-1.2, 81))
print ("Test 4 equals", sumFiniteGeoSeries(1, 342))
```

### Pr. 5: The *Lucas numbers* (Edouard Lucas 1842-1891) are similar to the Fibonacci sequence. They are generated by the recursive formula $ùêø_{ùëõ+2} = ùêø_{ùëõ+1} + ùêø_ùëõ , ~~~ùêø_1 = 2, ùêø_2 = 1$. Write a function that prints out the 1st $ùëõ$ Lucas numbers for a given $ùëõ$. Test your function for $ùëõ = 10, 25$.

```python
def lucasNumbers(n):
  lucn = []
  lucn.append(2) # Store L1 in the list
  lucn.append(1) # Store L2 in the list

  for i in range(2, n):
    lucn.append(lucn[i-2] + lucn[i-1])

  return lucn

n1 = 10
n2 = 25
print ("The first", n1, "Lucas numbers are \n", lucasNumbers(n1))
print ("The first", n2, "Lucas numbers are \n", lucasNumbers(n2))
```

### Pr. 6: The *Tribonacci sequence* is a generalization of the Fibonacci sequence and is given by $ùëá_{ùëõ+3} = ùëá_{ùëõ+2} + ùëá_{ùëõ+1} + ùëá_ùëõ , ~~~~ùëá_1 = 0, ùëá_2 = 0, ùëá_3 = 1 $. Write a function that prints out the 1st $ùëõ$ terms of the Tribonacci sequence for a given $ùëõ$. Test your function for $ùëõ = 15, 25$.

```python
def triboNacci(n):
  trib = []
  trib.append(0) # Store T1 in the list
  trib.append(0) # Store T2 in the list
  trib.append(1) # Store T3 in the list

  for i in range(3, n):
    trib.append(trib[i-3] + trib[i-2] + trib[i-1])

  return trib

n1 = 15
n2 = 25

print ("The first", n1, "Tribonacci numbers are \n", triboNacci(n1))
print ("The first", n2, "Tribonacci numbers are \n", triboNacci(n2))
```

### Pr. 8: Present Value of an Annuity: Calculating payments
#### Suppose we take out a car or a house loan and contract to pay this money back at a fixed interest rate in a fixed amount of time (Typically, 5 or 6 years for a car, or 30 years for a house). We want to compute our payment. The present value ùëÉùëâ is the initial loan amount and we pay it off to a future value of zero dollars.
#### The payment ùëÉùëÄùëá is given by the formula
$$ PMT = PV ~ \left[ \frac{i}{1-(1+i)^{-m}} \right] $$
#### The total amount paid to the bank is ùëÉùëÄùëá ‚àó ùëö.
#### The amount of interest that the bank earns is ùëÉùëÄùëá ‚àó ùëö ‚àí ùëÉùëâ.
<br>

#### Given ùëÉùëâ , ùëõ, ùëü, ùë°, write a function that computes: 1) the payment, 2) the total amount paid to the bank, and 3) the amount of interest the bank earns.
#### Round all numbers to two decimal places.
#### Optional: Save your function as a script.
#### Use your function to compute the payment and total amount of interest the bank earns for the following:

#### a) You take out a $25,000 loan for a car at 4% interest compounded monthly and plan to pay the loan off in 5 years.

#### b) You take out a $275,000 loan for a condo at 4% interest compounded monthly and plan to pay the loan off in 30 years.

#### c) You take out a $450,000 loan for a house at 5% interest compounded monthly and plan to pay the loan off in 30 years.


```python
def compPayment (pv, intr, yrs, n):
  i = intr/n
  m = yrs * n
  payment = np.round(pv * (i / (1-(1+i)**-m)), 2)
  return payment

def totalBankPayment (payment2, yrs2, n2):
  totalBankPayment = np.round(payment2 * (yrs2 * n2), 2)
  return totalBankPayment

def interestBankEarns (payment3, yrs3, pv2, n3):
  interestBankEarns = np.round(payment3 * (yrs3 * n3) - pv2, 2)
  return interestBankEarns

loanAmount1 = 25000
loanAmount2 = 275000
loanAmount3 = 450000

interestRate1 = .04
interestRate2 = .04
interestRate3 = .05

yearDuration1 = 5
yearDuration2 = 30
yearDuration3 = 30

compoundAmount = 12

print ("a) With a $", loanAmount1, "loan at", interestRate1, "% interest compounded monthly over", yearDuration1, "years, the monthly payment amount is $", compPayment(loanAmount1, interestRate1, yearDuration1, compoundAmount))
print ("The total amount paid to the bank is", totalBankPayment(compPayment(loanAmount1, interestRate1, yearDuration1, compoundAmount), yearDuration1, compoundAmount))
print ("The amount of interest the bank earns is", interestBankEarns(compPayment(loanAmount1, interestRate1, yearDuration1, compoundAmount), yearDuration1, loanAmount1, compoundAmount))

print ("b) With a $", loanAmount2, "loan at", interestRate2, "% interest compounded monthly over", yearDuration2, "years, the monthly payment amount is $", compPayment(loanAmount2, interestRate2, yearDuration2, compoundAmount))
print ("The total amount paid to the bank is", totalBankPayment(compPayment(loanAmount2, interestRate2, yearDuration2, compoundAmount), yearDuration2, compoundAmount))
print ("The amount of interest the bank earns is", interestBankEarns(compPayment(loanAmount2, interestRate2, yearDuration2, compoundAmount), yearDuration2, loanAmount2, compoundAmount))

print ("c) With a $", loanAmount3, "loan at", interestRate3, "% interest compounded monthly over", yearDuration3, "years, the monthly payment amount is $", compPayment(loanAmount3, interestRate3, yearDuration3, compoundAmount))
print ("The total amount paid to the bank is", totalBankPayment(compPayment(loanAmount3, interestRate3, yearDuration3, compoundAmount), yearDuration3, compoundAmount))
print ("The amount of interest the bank earns is", interestBankEarns(compPayment(loanAmount3, interestRate3, yearDuration3, compoundAmount), yearDuration3, loanAmount3, compoundAmount))
```
