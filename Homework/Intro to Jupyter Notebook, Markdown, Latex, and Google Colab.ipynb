### Pr. 5: Find the sum of the 1st 1000 positive integers by running the following three codes:

```python
n = 1000
mysum = 0
for i in range(1, n + 1):
    mysum = mysum + i
print("The sum of the first",n, "integers is", mysum)
```

```python
mysum = 0
for i in range(1, n + 1):
    mysum += i
print("The sum of the first",n, "integers is", mysum)
```

```python
print("The sum of the first",n, "integers is", sum(range(1, n + 1)))
```

### Pr. 6: Simulate the sum of 1 million throws of two six-sided dice and tabulate the results. Run the following code:

```python
SumList = []
SumCount = []
for i in range(1000000):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    SumList.append(die1 + die2)
for roll in range(2, 13):
    SumCount.append([roll, SumList.count(roll)])
print("[Roll, Number]")
for item in SumCount:
    print(item)
```
