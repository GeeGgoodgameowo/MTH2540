### Pr. 8: Write a function that computes the average absolute value $ \frac{|x| + |y|}{2} $.

####Test your function for following values of {ğ‘¥, ğ‘¦}:

```python
def avgAbs (x, y):
  absresult = (abs(x) + abs(y)) / 2
  return absresult

x1 = 7
y1 = 33
x2 = -42
y2 = 79
x3 = -17
y3 = -52

print("a) The average absolute value of {", x1, ",", y1, "} is", avgAbs(x1, y1))
print("b) The average absolute value of {", x2, ",", y2, "} is", avgAbs(x2, y2))
print("c) The average absolute value of {", x3, ",", y3, "} is", avgAbs(x3, y3))
```
