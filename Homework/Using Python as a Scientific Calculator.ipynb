### Pr. 2: Use Python to compute the following:
(For each part, first define an array such as x = np.array([2, 2.5, 3, 3.5, 4]).)

a) $2^x $ for $x = 2, 2.5, 3, 3.5, 4 $ <br>
b) $\ln x $ for $x = 10, 11, 12, 13, 14, 15$ <br>
c) $n! $ for $n = 5, 6, 7, 8, 9 , 10 $ <br>
d) $C(5,r) $ and $P(5,r) $ for $r = 0, 1, 2, 3, 4, 5$ <br>
e) $C(6,r) $ and $P(6,r) $ for $r = 0, 1, 2, 3, 4, 5, 6$


```python
from scipy.special import factorial, comb, perm

x = np.array ([2,2.5,3,3.5,4])
print ("a)", np.round(2**(x+1), 3))

xTwo = np.array ([10,11,12,13,14,15])
print ("b)", np.round(np.log(xTwo), 3))

xThree = np.arange(5,11)
print ("c)", np.round(factorial(xThree, exact=True), 3))

xFour = np.arange(6)
print ("d)", np.round(comb(5,xFour), 3), np.round(perm(5,xFour), 3))

xFive = np.arange(7)
print ("e)", np.round(comb(6,xFive), 3), np.round(perm(6,xFive), 3))
```

a) $3x^2+2x-1=0 $<br>

### Pr. 10: How can you use // to return all but the units digit of a given integer $n$? Use this process to have Python return all but the units digit for $n$ = 1234, $n$ = 517, and $n$ = 3141592653589.

```python
print("// 10 would give all but the remainder, i.e units digit, of any integer.")

nOne = 1234
nTwo = 517
nThree = 3141592653589

print(nOne // 10)
print(nTwo // 10)
print(nThree // 10)
```
